```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('s5092_NO2_calibration.csv')
```

```python
df['dt_time'] = pd.to_datetime(df['dt_time'])
df.set_index(df["dt_time"])
```

```python
#80:20 split
training_set = df[df['dt_time'].between('2023-04-01', '2023-04-25 23:50:00')]
testing_set = df[df['dt_time'].between('2023-04-26', '2023-05-04')]
#70:30 split
training_set1 = df[df['dt_time'].between("2023-04-01 12:30:00","2023-04-22 21:00:00")]
testing_set1 = df[df["dt_time"].between("2023-04-22 21:15:00", "2023-05-04")]

X_train = training_set.drop(["dt_time", "NO2"], axis = 1)
X_test = testing_set.drop(['dt_time', 'NO2'], axis = 1)
y_train = training_set["NO2"]
y_test = testing_set['NO2']
X_train1 = training_set1.drop(["dt_time", "NO2"], axis = 1)
y_train1 = training_set1["NO2"]
X_test1 = testing_set1.drop(["dt_time", "NO2"], axis = 1)
y_test1 = testing_set1["NO2"]
```

```python
#(80:20 Split) Linear regression model on trainig data
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
#making predictions
y_pred = lr.predict(X_train)
```

```python
#Visualise our data
plt.scatter(y_train, y_pred, color = 'r', label = 'Training Data')
plt.title('80:20 split')
plt.xlabel('Ground Truth NO2')
plt.ylabel('Predicted NO2')
plt.legend()
plt.show()
```

```python
#R2 Score from training data
from sklearn.metrics import r2_score
print(f'r2 score from training data: {r2_score(y_train, y_pred)}')
```

```python
#Testing data
y_pred_test = lr.predict(X_test)
```

```python
#Visualise our data
plt.scatter(y_test, y_pred_test, color = 'r', label = 'Testing Data')
plt.title('80:20 split')
plt.xlabel('Ground Truth NO2')
plt.ylabel('Predicted NO2')
plt.legend()
plt.show()
```

```python
#R2 Score from testing data
from sklearn.metrics import r2_score
print(f'r2 score from testing data: {r2_score(y_test, y_pred_test)}')
```
