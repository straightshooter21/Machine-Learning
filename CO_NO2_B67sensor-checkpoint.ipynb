```python
#imports
from statsmodels.tsa.seasonal import seasonal_decompose
import pandas as pd
import numpy as np
import chart_studio.plotly as py
import cufflinks as cf
import seaborn as sns
import plotly.express as px
%matplotlib inline

from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected = True)
cf.go_offline()
import matplotlib.pyplot as plt
```

```python
#reading our data in
df = pd.read_csv('data_4C752509EF48.csv')
df['dt_time'] = pd.to_datetime(df['dt_time'])
df.set_index(df['dt_time'])
NO2df = df.drop(['CO', 'coop1', 'coop2'], axis = 1)
COdf = df.drop(['NO2', 'no2op1', 'no2op2'], axis = 1)
NO2df = NO2df.dropna()
COdf = COdf.dropna()
pd.set_option('display.max_rows', None)
COdf.shape
```

```python
#Seasonal decomposition of NO2 data
resultNO2 = seasonal_decompose(NO2df['NO2'].to_numpy(), model='additive', period=24*4)  # Period of 12 for monthly data
resultNO2.plot()
plt.show()
periodic_estimate = resultNO2.seasonal
residual = resultNO2.resid
residual[np.isnan(residual)] = 0
resultNO2final = periodic_estimate + residual
resultNO2final = pd.DataFrame(resultNO2final, columns=['NO2'])
```

```python
#NO2 - 70:30 split
training_setNO2 = NO2df[NO2df['dt_time'].between('2023-04-13 00:15:00', '2023-06-01 17:15:00')]
testing_setNO2 = NO2df[NO2df['dt_time'].between('2023-06-01 17:30:00', '2023-06-10')]
```

```python
#NO2 Creating feature vector and label with 70:30 split
X_trainNO2 = training_setNO2.drop(['dt_time', 'NO2'], axis = 1)
X_testNO2 = testing_setNO2.drop(['dt_time', 'NO2'], axis = 1)

y_trainNO2 = resultNO2final[:1878]
y_testNO2 = resultNO2final[1878:]
```

```python
#NO2: importing linear regression and training
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_trainNO2, y_trainNO2)
```

```python
#NO2: Making predictions using training data and plotting for training data
y_pred_trainNO2 =lr.predict(X_trainNO2)
y_pred_trainNO2 = pd.DataFrame(y_pred_trainNO2, columns = ['Predicted NO2'])
training_pred_vs_gtNO2 = pd.concat([y_pred_trainNO2, y_trainNO2], axis = 1)
fig = px.scatter(training_pred_vs_gtNO2, x="NO2", y="Predicted NO2",color_discrete_sequence = ['red'], 
                 title = 'NO2 Training Data Scatter Plot')
fig.show()
```

```python
#NO2: Making predictions using testing data and plotting for testing data
y_pred_testNO2 = lr.predict(X_testNO2)
y_pred_testNO2 = pd.DataFrame(y_pred_testNO2, columns = ['Predicted NO2'])
testing_pred_vs_gtNO2 = pd.concat([y_pred_testNO2, y_testNO2.reset_index()], axis = 1)
fig = px.scatter(testing_pred_vs_gtNO2, x="NO2", y="Predicted NO2",color_discrete_sequence = ['red'], 
                 title = 'NO2 Testing Data Scatter Plot')
fig.show()
```

```python
#NO2: r2 scores from seasonal component
from sklearn.metrics import r2_score
print('For NO2')
print(f'r2 score from training data: {r2_score(y_trainNO2, y_pred_trainNO2)}')
print(f'r2 score from testing data: {r2_score(y_testNO2, y_pred_testNO2)}')
```

```python
#NO2: Time Series
y_pred_trainNO2 = pd.DataFrame(y_pred_trainNO2, columns = ['Predicted NO2'])
dt_time = training_setNO2['dt_time']
df_timeseriesNO2training = pd.concat([y_pred_trainNO2.reset_index(), dt_time.reset_index(), y_trainNO2.reset_index()], axis=1)
px.line(df_timeseriesNO2training, x='dt_time', y = ['NO2', 'Predicted NO2'], 
        labels = {'x': 'Date', 'y': 'NO2'}, title = 'NO2 Training Data: Time Series Plot', color_discrete_sequence = ['red', 'blue'])
```

```python
#Time Series Plot for testing data for NO2
y_pred_testNO2 = pd.DataFrame(y_pred_testNO2, columns = ['Predicted NO2'])
dt_time1 = testing_setNO2['dt_time']
df_timeseriesNO2testing = pd.concat([y_pred_testNO2.reset_index(), dt_time.reset_index(), y_testNO2.reset_index()], axis=1)
px.line(df_timeseriesNO2testing, x='dt_time', y = ['NO2', 'Predicted NO2'], 
        labels = {'x': 'Date', 'y': 'NO2'}, title = 'NO2 Testing Data: Time Series Plot', color_discrete_sequence = ['red', 'blue'])
```
