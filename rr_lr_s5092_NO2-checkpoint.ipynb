```python
#imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Read in our data
df = pd.read_csv('s5092_NO2_calibration.csv')
df
```

```python
#set dt_time as index
df['dt_time'] = pd.to_datetime(df['dt_time'])
df.set_index(df['dt_time'])
```

```python
#Splitting our data (70-30)
training_data = df[df['dt_time'].between("2023-04-01 12:30:00","2023-04-22 21:00:00")]
testing_data = df[df["dt_time"].between("2023-04-22 21:15:00", "2023-05-04")]
X_train = training_data.drop(["dt_time", "NO2"], axis=1)
X_test = testing_data.drop(['dt_time', "NO2"], axis=1)
y_train = training_data['NO2']
y_test = testing_data['NO2']
```

```python
#Applying Ridge Regression on training data
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha = 50, max_iter = 100, tol = 0.1)
ridge_reg.fit(X_train, y_train)
```

```python
#making predictions
y_pred_train = ridge_reg.predict(X_train)
```

```python
#Visualising our data
plt.scatter(y_train, y_pred_train, label = 'Training Data', color = 'r')
plt.xlabel('Ground Truth NO2')
plt.ylabel('Predicted NO2')
plt.show
```

```python
#Time-Series Plot
fig, ax = plt.subplots()
ax.plot(training_data['dt_time'], y_train, label = 'Ground Truth Values', color = 'r')
ax.set_xlabel('Time')
ax.set_ylabel('Ground Truth NO2')
ax1 = ax.twinx()
ax1.plot(training_data['dt_time'], y_pred_train, label = "Predicted Values", color = 'b')
ax1.set_xlabel("Time")
ax1.set_ylabel("Predicted NO2")
fig.legend()
plt.show()
```

```python
#imoorting r2_score
from sklearn.metrics import r2_score
print(f'r2 score from training data: {r2_score(y_train, y_pred_train)}')
```

```python
#Testing Data
ridge_reg.fit(X_test, y_test)
```

```python
#Making predictions 
y_pred_test = ridge_reg.predict(X_test)
```

```python
plt.scatter(y_test, y_pred_test, label = 'Testing Data', color = 'r')
plt.xlabel('Ground Truth NO2')
plt.ylabel('Predicted NO2')
plt.show()
```

```python
print(f'r2 score from testing data: {r2_score(y_test, y_pred_test)}')
```

```python
#Time Series plot
fig1,ax2 = plt.subplots()
ax2.plot(testing_data['dt_time'], y_test, label = 'Ground Truth Values', color = 'r')
ax2.set_xlabel("Time")
ax2.set_ylabel("Ground Truth NO2")
ax3 = ax2.twinx()
ax3.plot(testing_data['dt_time'], y_pred_test, label = 'Predicted values', color = 'b')
ax3.set_xlabel('Time')
ax3.set_ylabel("Predicted NO2")
fig.legend()
plt.show
```
